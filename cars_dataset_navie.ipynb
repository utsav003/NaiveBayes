```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
col = ['buying','maint','doors','persons','lug_boot','safety','Class_Values']
data1 = pd.read_csv('C:\\Users\\UTSAV\\Desktop\\MY_Practice\\car.data',header= 0 , names = col,index_col = None)
data1.head(2)
```

```python
data = data1
```

```python
data.nunique()
data.safety.unique()
```

```python
var = preprocessing.LabelEncoder()
buying = var.fit_transform(data1.buying)
maint = var.fit_transform(data1.maint)
lug_boot = var.fit_transform(data1.lug_boot)
safety = var.fit_transform(data1.safety)
Class_Values = var.fit_transform(data1.Class_Values)
```

```python
#len(data.doors)
for i in range(len(data.doors)):
    if(data['doors'][i] == '5more'):
        data['doors'][i] = 5
    elif(data['doors'][i] == '2'):
        data['doors'][i] = 2 
    elif(data['doors'][i] == '3'):
        data['doors'][i] = 3
    elif(data['doors'][i] == '3'):
        data['doors'][i] = 3
    elif(data['doors'][i] == '4'):
        data['doors'][i] = 4
```

```python
for i in range(len(data.persons)):
    if(data['persons'][i] == 'more'):
        data['persons'][i] = 5
    elif(data['persons'][i] == '2'):
        data['persons'][i] = 2 
    elif(data['persons'][i] == '3'):
        data['persons'][i] = 3
    elif(data['persons'][i] == '3'):
        data['persons'][i] = 3
    elif(data['persons'][i] == '4'):
        data['persons'][i] = 4
```

```python
features = list(zip(buying,maint,data.doors,data.persons,lug_boot,safety))
features
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(features,Class_Values)
predicted= model.predict([[3,3,0,0,2,1]])
print("Predicted Value:", predicted)
```

```python
print(Class_Values)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, Class_Values, test_size = 0.09,random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```
